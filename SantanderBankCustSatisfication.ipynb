```python
result=pd.DataFrame()
def UDF_ttest(df):
    from scipy import stats
    global result
    col_b4=df.shape[1]
    for col in df.iloc[:,1:]:
        Unhappy=df[col][df.TARGET==1] 
        Happy=df[col][df.TARGET==0] 
        tstat,pvalue=stats.ttest_ind(Happy,Unhappy)
        print("\n",df[col].name, "p-value: ", pvalue)
        if (pvalue > 0.05):
            print("DROPPED")
            df=df.drop(col,axis=1)
        else:
            print("Retained")
    print("No of Columns Retained: ",df.shape[1])
    print("No of Columns Dropped: ",col_b4-df.shape[1])
    result=df;
            
```

```python
UDF_ttest(df)
```
